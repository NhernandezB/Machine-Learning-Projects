<h1>Linear Regression</h1>

```python
# Import libraries 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data sample
X_normal = np.array([1, 2, 3, 4, 5])
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])
```

```python
# Create linear regression model
model = LinearRegression()
```

```python
# Fitting data with the model
model.fit(X, y)
```

```python
# Make a prediction
x_prediction = np.array([6]).reshape(-1, 1)
y_prediction = model.predict(x_prediction)
```

```python
# Data Visualization 
plt.scatter(X, y, color = 'g', label = 'Data')
plt.plot(X, model.predict(X), color='r', label='Linear Regression')
plt.xlabel('Indendient Variable')
plt.ylabel('Dependent Variable')
plt.legend()
```

```python
print("Slope (Coefficient):", model.coef_[0])
print("Intersection (Ordered to Origin):", model.intercept_)
```
